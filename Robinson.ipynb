```python
# Các mệnh đề CNF
clauses = [
    {"~F", "~B", "Z"},     # (F ∧ B) -> Z
    {"~C", "~D", "F"},     # (C ∧ D) -> F
    {"~A", "D"},           # A -> D
    {"A"},                 # fact
    {"E"},                 # fact
    {"B"},                 # fact
    {"C"},                 # fact
    {"~Z"}                 # phủ định kết luận
]

# Hàm lấy phủ định
def negate(literal):
    if literal.startswith("~"):
        return literal[1:]
    return "~" + literal

# Thuật toán Robinson Resolution
new = set()

print("Các clause ban đầu:")
for c in clauses:
    print(c)

while True:
    n = len(clauses)

    pairs = [
        (clauses[i], clauses[j])
        for i in range(n)
        for j in range(i + 1, n)
    ]

    resolved = False

    for (ci, cj) in pairs:
        for literal in ci:
            if negate(literal) in cj:

                resolvent = (ci - {literal}) | (cj - {negate(literal)})

                print("\nResolve:")
                print(ci, "với", cj)
                print("=>", resolvent)

                if len(resolvent) == 0:
                    print("\nSuy ra mệnh đề rỗng {}")
                    print("=> Chứng minh được Z")
                    resolved = True
                    break

                if frozenset(resolvent) not in new:
                    new.add(frozenset(resolvent))
                    clauses.append(resolvent)

        if resolved:
            break

    if resolved:
        break

    if all(frozenset(c) in new for c in clauses):
        print("Không suy ra được.")
        break
```
